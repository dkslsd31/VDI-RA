```python
import pandas as pd
data = pd.read_csv('/content/best-selling-books.csv')
data.head()
```

```python
print(data.info())
```

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(data['Original language'], data['Approximate sales in millions'])
plt.xlabel('Original language')
plt.ylabel('Approximate Sales in Millions')
plt.title('Approximate Sales in Millions by language')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability if needed
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
from matplotlib import pyplot as plt
data['Approximate sales in millions'].plot(kind='hist', bins=20, title='Approximate sales in millions')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
genre_sales = data.groupby('Genre')['Approximate sales in millions'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(50, 50))
plt.barh(genre_sales['Genre'], genre_sales['Approximate sales in millions'], color='lightcoral')
plt.xlabel('Approximate Sales (in millions)')
plt.title('Total Sales by Genre for Top-Selling Books')
# Menampilkan plot
plt.tight_layout()
plt.show()
```
